```python
schools_na = (schools_df.isnull().sum() / len(schools_df)) * 100
schools_na = schools_na.drop(schools_na[schools_na == 0].index).sort_values(ascending=False)[:30]
schools_missing_data = pd.DataFrame({'Missing Ratio':schools_na})
schools_missing_data
```

```python
# Percent missing data by feature

f, ax = plt.subplots(figsize=(15, 12))
plt.xticks(rotation='90')
sns.barplot(x=schools_na.index, y=schools_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
```
