We have lots of columns to work with, we probably dont need most of these for our purposes of creating a feature for our SE data.

To make things easier we will create a new column named addr_zip, the same as in our SE data for any grouping we may need to do.

```python
schools_df['addr_zip'] = schools_df['location'].str.extract(r'(\d{5}\-?\d{0,4})')
```

Let's narrow down the columns that we are going to work with.

we are choosing the ones we feel are most important when it comes to choosing the best school in an area like its graduation_rate, location, and others.

```python
columns = ['school_name','overview_paragraph','neighborhood','location',
           'website','grades2018','finalgrades','total_students','graduation_rate','attendance_rate','college_career_rate','Borough','addr_zip','Latitude',
           'Longitude']

final_df = schools_df[columns]
```

Now that we've narrowed down our dataset lets look for missing values 

```python
final_na = (final_df.isnull().sum() / len(final_df)) * 100
final_na = final_na.drop(final_na[final_na == 0].index).sort_values(ascending=False)[:30]
schools_missing_data = pd.DataFrame({'Missing Ratio':final_na})
schools_missing_data
```

## imputing missing values

We will be imputing all the missing values with the mean as it makes the most sense currently

```python
final_df['graduation_rate'].fillna(final_df['graduation_rate'].mean(),inplace=True)
final_df['college_career_rate'].fillna(final_df['college_career_rate'].mean(),inplace=True)
final_df['attendance_rate'].fillna(final_df['attendance_rate'].mean(),inplace=True)
```

Check again for any left over missing values and look at our new dataframe

```python
final_na = (final_df.isnull().sum() / len(final_df)) * 100
final_na = final_na.drop(final_na[final_na == 0].index).sort_values(ascending=False)[:30]
schools_missing_data = pd.DataFrame({'Missing Ratio':final_na})
schools_missing_data
```

```python
final_df
```

Lets look at the column's we feel best show how good a school is the graudation rate and college career rate

```python
final_df.plot(kind='scatter',x='graduation_rate',y='college_career_rate')
```

we can see that there exist a positive correlation between graduation rate and college_career_rate and thus we will only use college career rate as a measure of a good high school.

Lets create a new dataframe with the best school, defined by the college_career_rate, in a given neighborhood.

```python
best_schools_neighborhood_df = final_df[final_df['college_career_rate'] == final_df.groupby('neighborhood')['college_career_rate'].transform('max')].reset_index(drop=True)
```
